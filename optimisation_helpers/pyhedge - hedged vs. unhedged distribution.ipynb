# pyhedge - hedged vs. unhedged distribution

Chosing to hedge comes down your return expectation and the confidence you have in those expectations. Quantifying that certainty is difficult by can be done through generating a distribution of returns. With this distribution of returns, we can then measure the impact hedging has on the expectancy.

hedged_er assumes returns will be lognormally distribution, this is a very naive assumption. Future implementations will include more appropriate assumptions for the distribution of returns.

---

## Import external libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Import pyport

```python
import pyport.porthedge as hdge
```

## Define parameters for simulation

```python
mer = 0.05
vol = 0.15
hedge_cost = 0.015
hedge_strategy = 'put'
leg1_strike = -0.1
```

## 

```python
res = hdge.hedged_er(0.05, 0.15, 0.015, hedge_strategy='put', leg1_strike=-0.1)
un, hg = res['unhedged'], res['hedged']
```

```python
un.plot(kind='hist', bins=100, figsize=(15, 8), label='unhedged', color='b', alpha=0.75)
hg.plot(kind='hist', bins=100, label='hedged', color='g', alpha=0.75)
plt.axvline(un.mean(), color='b', linewidth=1)
plt.axvline(hg.mean(), color='g', linewidth=1)
plt.title('Unhedged vs. Hedged Simulated Distribution', fontweight='bold', fontsize=18)
plt.legend(loc='best')
plt.xlim(-0.5, 0.5)
plt.tight_layout();
```
